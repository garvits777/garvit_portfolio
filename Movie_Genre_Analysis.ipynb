```python
import pandas as pd
```

```python
movies = pd.read_csv(r'G:\DA\Alex\Pandas\Files\imdb_movies.csv')
movies.head(5)
```

```python
pd.set_option('display.max.rows',11000)
pd.set_option('display.max.columns',22)
```

```python
movies.head()
```

```python
# Finding the duplicates in the data

movies[movies.duplicated()]
```

```python
movies.info()
```

```python
movies.drop_duplicates(inplace = True)
```

```python
movies.dropna(subset = ['genres'],inplace = True)
```

```python
movies['profit'] = movies['revenue'] - movies['budget']
```

```python
movies_genre = movies[['popularity','budget','revenue','original_title','runtime','genres','release_year','vote_count','vote_average','profit']]
```

```python
movies_genre.head()
```

```python
from pandas import Series, DataFrame
```

```python
from pandas import Series, DataFrame

# Assuming movie_genres is a DataFrame with a 'genres' column
split = movies_genre['genres'].str.split('|').apply(Series, 1).stack()
split.index = split.index.droplevel(-1)

# To display the results
split.name = 'genres_split'
del movies_genre['genres']
movies_genre = movies_genre.join(split)

```

```python
movies_genre.head()
```

```python
type(movies_genre)
```

```python
genres_count = pd.DataFrame(movies_genre.groupby('genres_split').original_title.nunique()).sort_values('original_title', ascending=True)
```

```python
genres_count['original_title'].plot.pie(title = 'Movies Per Genre %',autopct = '%1.1f%%',figsize = (10,10))
```

```python
genres_count['original_title'].plot.barh('Movies Per Genre',color = 'DarkBlue',figsize = (10,10))
```

```python
genres_avg = movies_genre.groupby('genres_split').mean()
pd.options.display.float_format = '{:2f}'.format
genres_avg
```

```python
genres_avg
```

```python
genres_avg.sort_values('budget',ascending=True, inplace = True)
```

```python
genres_avg[['budget', 'revenue']].plot.barh(
    color=['DarkBlue', 'c'], 
    figsize=(10, 10), 
    title='Budget and Revenue For Each Genre'
)
```

```python
genres_avg.sort_values('profit',ascending=True, inplace = True)
```

```python
genres_avg['profit'].plot.barh(
    color='DarkBlue', 
    figsize=(10, 10), 
    title='Profit By Genre'
)
```

```python
genres_avg.sort_values('popularity',ascending=True, inplace = True)
```

```python
genres_avg['popularity'].plot.barh(
    color='DarkBlue', 
    figsize=(10, 10), 
    title='Popularity By Genre'
)
```
